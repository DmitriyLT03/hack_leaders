```python
import pandas as pd
```

```python
dataset = pd.read_csv('./full_dataset.csv', sep='\t')
```

```python
dataset
```

```python
def create_subtable(data, napr: str, tnved, region=None):
    if region == None: 
        return data[(data.napr == napr) & (data.tnved == tnved)]
    return data[(data.napr == napr) & (data.tnved == tnved) & (data.Region == region)]
```

```python
test_df = create_subtable(dataset, "ИМ", 1904)
test_df.Stoim = test_df.Stoim.apply(lambda x: x.replace(',','.')).astype('float32')
```

```python
test_df_group = pd.DataFrame(test_df.groupby(['period', 'nastranapr']).sum()['Stoim']).reset_index()
```

```python
import matplotlib.pyplot as plt
test_2df = create_subtable(dataset, "ИМ", 1904)
test_2df.Stoim = test_2df.Stoim.apply(lambda x: x.replace(',','.')).astype('float32')
test_2df = test_2df[test_2df.nastranapr == 'PL']
test_2df = pd.DataFrame(test_2df.groupby(['period']).sum()['Stoim']).reset_index()
test_2df.period = test_2df.period.index.values.astype('datetime64[M]').astype(int)
test_2df = test_2df.sort_values(by=['period'], ascending=True)
test_2df.Stoim.reset_index(drop = True).plot(style='-')
```

```python
test_2df = create_subtable(dataset, "ИМ", 1904)
test_2df.Stoim = test_2df.Stoim.apply(lambda x: x.replace(',','.')).astype('float32')
test_2df = test_2df[test_2df.nastranapr == 'NG']
test_2df = pd.DataFrame(test_2df.groupby(['period']).sum()['Stoim']).reset_index()
test_2df.period = test_2df.period.index.values.astype('datetime64[M]').astype(int)
test_2df = test_2df.sort_values(by=['period'], ascending=True)
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(test_2df.period.values.reshape(-1, 1), test_2df.Stoim.values)
plt.plot(test_2df.period.values, reg.predict(test_2df.period.values.reshape(-1, 1)))
test_2df.Stoim.reset_index(drop = True).plot(style='-')
```
